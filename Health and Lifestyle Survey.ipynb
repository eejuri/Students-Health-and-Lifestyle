## 1. Introduction

### 1.1. Topic overview

I undertook a survey on health and lifestyle habits of the University of London (hereafter - UoL) MSc Data Science students, which was based on the survey project specification [1]. The survey's main finding was that the majority of UoL MSc Data Science students pursued a relatively healthy lifestyle, apart from having sleep and mental health issues. My report has the following sections: Introduction, Research topic, Data sources, Data overview and pre-processing, Analysis and results, Conclusion and Evaluation.

### 1.2. Background and context

As referenced in my survey project specification [1], past research revealed that “a healthier lifestyle is consistently associated with a longer life expectancy” [2] and that “healthy lifestyle behaviours are positively and independently associated with academic achievement” [3]. Furthermore, various studies looked into health and lifestyle behaviours of university students [1]. For example, Schmidt [4] investigated the impact of socio-demographic factors on Swedish university students’ lifestyle behaviours (physical activity, perceived stress and eating behaviours) and self-rated health. She found that both lifestyle behaviours and self-rated health could be to some extent predicted by socio-demographic factors like gender, mother tongue and parents’ educational level. Bennasar-Veny et al. [5] assessed the prevalence, interrelation and clustering of health-related lifestyle habits (smoking, alcohol consumption, physical activity and diet quality) among students from the University of the Balearic Islands. The analysis indicated the presence of three distinct groups: unhealthy lifestyle habits with moderate risk, unhealthy lifestyle habits with high risk; and healthy lifestyle habits with low risk. Similarly, Dodd et al. [6] looked into the prevalence and clustering of lifestyle risk factors within a UK Higher Education institution. Three subgroups emerged from the cluster analysis: ‘unhealthy/high risk’, ‘moderately healthy/moderate risk’ and ‘healthy/low risk’ group. Finally, Deniozou [7] in her survey assessed a number of health behaviours and lifestyle features of undergraduate and postgraduate research students at the University of Edinburgh. While overall results looked encouraging, she proposed interventions in the areas of physical activity, diet, sexual health, alcohol consumption and drugs.

### 1.3. Motivation and objectives

As I pointed out in my survey project specification [1], I try to pursue a healthy lifestyle as it helps me live a happier life. My research objectives were to explore and describe health and lifestyle habits of the UoL MSc Data Science students. In my study I replicated part of the health and lifestyle survey at the University of Edinburgh by Deniozou [7] and compared my results with that survey. Based on the results I made recommendations on further research that could potentially improve the lifestyle of UoL students (please see section Conclusion).

## 2. Research  topic

### 2.1. Research questions

Deniozou [7] developed an online survey that asked demographic questions as well as questions about the physical activity, diet, sleep, sexual health, alcohol consumption, smoking, drug usage and mental health of the University of Edinburgh students. I asked almost identical survey questions but in a different order: a student’s awareness of her/his sleep, diet, physical and mental health, alcohol and tobacco consumption habits and demographic questions. Furthermore, as we were advised by the survey project specification instructions to aim for three or four specific research questions, I decided to omit questions on sexual health and drugs that were probably more sensitive than others and were likely to be more relevant to undergraduate than postgraduate students. 1. How healthy is a student’s sleep?
2. How healthy is a student’s diet?
3. What is a student’s awareness of her/his physical and mental health?

### 2.2. Population and sampling methodology

My population was all UoL MSc Data Science students (globally) and my sample was the 40 UoL MSc Data Science students that responded to my survey questions. Deniozou [7] used stratified sampling by dividing a population into subclasses with common characteristics. As I did not get enough responses from each subclass, I used simple random sampling - the randomised selection of a small segment of individuals from an entire population, providing each individual with an equal and fair probability of being chosen that allows for unbiased data collection [8]. Some bias came from imbalanced categories (please see sub-section My survey of section Data collection / Survey design) and that I could only lessen to some extent (please see section Data overview and pre-processing). 

### 2.3. Domain concepts

I used domain concepts from Deniozou [7], Nishat [9] and MentalHealth.gov [10].
* “One portion is approximately one piece of medium-sized fruit or 80g of any one vegetable” [7].
* “Drinks may include water, milk, fruit juices, tea and coffee but not alcoholic, sugary, soft and fizzy drinks” [7].
* Physical health is the well-being and the proper functioning of the body [9].
* Mental health comprises “our emotional, psychological and social well-being” [10].
* One alcohol unit is: “half pint of 3.6% beer, half a small glass of 12% wine, one single measure of 40% spirits” [7].
* Domicile on entry is an area of permanent residence before beginning studies at the university [7].

## 3. Data sources

### 3.1. My survey

My survey (file UoL survey.xlsx attached) was conducted between July 16 and August 5 2022 with help of Online surveys, the online survey tool designed for Academic Research, Education and Public Sector organisations (www.onlinesurveys.ac.uk). I reached out to students using Slack and WhatsApp channels as well as the online forum. I received 40 responses with some imbalanced categories: for instance, 31 men but only 9 women, 14 students from England but only 1 from Europe outside EU/UK/Channel Islands/Isle of Man and 17 students aged 35 or over but only 3 aged 21-24. As imbalanced data sets typically generate biased results [11], they should be treated with particular care. Another limitation was that as in Deniozou [7], some survey questions involved self-assessment with the disadvantage of possible social desirability bias (please see sub-section External data below).

As noted in section Research topic, I used broadly the same survey questions as Deniozou [7]. However, there were several differences [1]. First, we were advised by the survey project specification instructions to aim for approximately ten survey questions, in addition to basic demographic questions. Hence from 39 survey questions I chose ten, in my view, potentially most informative ones, plus three demographic questions. In particular, I decided to omit questions on sexual health and drugs (please see sub-section Research questions for reasoning). 

Second, I changed the survey structure. In line with survey design recommendations by Forth [12], I decided to start with easier questions on sleep and diet, followed by more cognitively demanding questions on physical/mental health and potentially more sensitive questions on alcohol and tobacco consumption, and finished with easier demographic questions. Furthermore, I merged the questions on physical and mental health into a grid question 6 as they have the same measurement scale.

Finally, I changed the categories of the demographic questions (number 11-13). In particular, in the Age question I started from the category “20 or under” instead of “17 or under” and ended with the category “35 or over” instead of “25 or over” to reflect the likely older age of the student population and to avoid a reference to an age below 18. In addition, in the Domicile question I replaced the “Other UK” category with categories “England”, “Wales” and “Northern Ireland” to include all the UK constituent countries, and replaced the “Overseas” category with categories “Europe outside EU/UK/Channel Islands/Isle of Man”, “Asia”, “Americas”, “Africa” and “Australia/Oceania” to make the analysis potentially more interesting. Furthermore, in each of the demographic questions I added a category “Prefer not to say” to give respondents an option to protect their privacy. 

### 3.2. External data

I found the external survey [7] on Internet and copied the results relevant to my research questions and demographic questions into an Excel spreadsheet (please find Edinburgh survey.xlsx file attached). The research was conducted as part of the Healthy University Project, which aimed to actively promote and deliver tangible health and wellbeing benefits for the University of Edinburgh community with a particular focus on students’ health. A total of 1189 undergraduate and postgraduate research students (15% response rate) participated in the study between the end of April and the end of May 2015 by email using SurveyMonkey, an online cloud-based survey development tool. The results were visualised in charts and tables in percentage terms rather than in raw numbers. The data source is public, so for me there should be no issues with using the data.

The study had limitations. First, due to prolonged delays with ensuring ethical approval from the University’s Ethics committee, the survey was released later than planned and the students were asked to participate during the revision and examination period, which possibly affected their normal health and lifestyle habits. Second, postgraduate taught students were not surveyed due to clashes with other surveys running during the same time period. Third, some survey questions involved self-assessment with the disadvantage of possible social desirability bias, according to which survey respondents may provide answers in a manner viewed more favourably by others. Finally, data were not analysed by demographic variables or socio-economic data.

## 4. Data overview and pre-processing

The concept of tidy data was established by Hadley Wickham in his paper [13] in 2014. According to Wickham, tidy datasets "have a specific structure: each variable is a column, each observation is a row, and each type of observational unit is a table". I tried to follow that structure.

Overall I had 14 variables in my UoL survey:

1.	A student’s self-assessment of her/his sleep quality (ordinal data).
2.	A student’s hours of sleep on an average night (ordinal data).
3.	A student’s self-assessment of her/his diet healthiness (ordinal data).
4.	Portions of fruit and vegetables eaten by a student on a typical day (ordinal data).
5.	The number of 200ml glasses of fluids drunk by a student on an average day (ordinal data).
6.	A student’s self-assessment of her/his physical health (ordinal data).
7.	A student’s self-assessment of her/his mental health (ordinal data).
8.	The average number of alcohol units consumed by a student per week (ratio data).
9.	The frequency of having 6(if female)/8(if male) or more alcohol units by a student on a single occasion, or excess alcohol consumption frequency (ordinal data).
10.	A student’s self-assessment of her/his smoking habit (ordinal data).
11.	The number of cigarettes smoked by a student over the past week (ordinal data).
12.	A student’s age (ordinal data).
13.	A student’s gender (nominal data).
14.	A student’s domicile on entry (nominal data).

To clean the data, first I listed all categories for categorical variables and replaced column names with shorter ones. Second, I set ordinal and nominal categorical variables and dropped the health column with NaN values. I also found that there was another NaN value (missing response) in the excess alcohol consumption frequency data. Yet I dealt with it later, excluding the relevant row with help of iloc and np.r_ functions, when counting the variable's categories in percentage terms and creating a normalised table.

Third, I counted categories' unique values, with the maximum count corresponding to the mode category, and deleted the values with zero counts. Fourth, I merged category "EU" with category "Europe outside EU/UK/Channel Islands/Isle of Man" that had only one response and was a potential source of bias, to make a category "Europe outside the UK". Note that England was the only UK country represented in my survey, with no respondents from Scotland, Wales or Northern Ireland.

Sixth, I counted the categories in percentage terms and created normalised tables for the first 11 variables in relation to the demographic variables.

Seventh, I created 13 dataframes with the above variables (except domicile, as the UoL and the University of Edinburgh had completely different categories) in percentage terms at both universities, effectively making University a nominal variable with two categories, and set ordinal and nominal categorical variables for each dataframe. Based on these dataframes I created pivot tables representing the variables at both universities. Finally, I calculated correlations between the two universities by categories.

## 6. Conclusion

In recalling my research questions, which were:

1. How healthy is a student’s sleep?
2. How healthy is a student’s diet?
3. What is a student’s awareness of her/his physical and mental health?
4. What are a student’s alcohol and tobacco consumption habits?

I think that I managed to answer all the research questions in my survey.

The surveyed UoL students on average had a worse self-assessment of their sleep quality and reported less hours of sleep than the University of Edinburgh respondents, which can be related to the differences in age and gender distributions. Both the surveyed younger and female students at the UoL on average reported better sleep quality than their coursemates while UoL female students reported more hours of sleep than their male coursemates. 

Over 90% of both university respondents reported average, healthy or very healthy diet while on average, the surveyed UoL students revealed smaller consumption of fruit and vegetables but bigger intake of fluids than the University of Edinburgh respondents. 30-year-olds and older UoL students on average reported healthier diets and more fruit and vegetables consumption than younger categories.

80-84% of the surveyed students at both universities rated physical health as good, very good or excellent, with the UoL 21-24-year-olds on average rating it higher than their older coursemates. However, the surveyed UoL students had a much worse self-assessment of their mental health than the University of Edinburgh respondents on average. Worryingly, 37.5% of the surveyed African students at the UoL rated it as poor, indicating that this category may need additional support.

On average, excess alcohol consumption seemed to be less of an issue for the surveyed UoL students than for the University of Edinburgh respondents, with the UoL respondents consuming only about half of alcohol drunk by the surveyed University of Edinburgh students per week. Only 10% of the UoL respondents consumed more than 9 alcohol units weekly, all of them aged 30 or older.

Finally, my survey showed that the majority of both university students had never smoked and that there was a much smaller share of heavy smokers at the UoL compared to the University of Edinburgh survey.

Overall, my study found that the majority of UoL MSc Data Science students pursued a relatively healthy lifestyle, apart from having sleep and mental health issues. I believe that these issues deserve deeper research that potentially can improve the lifestyle of UoL students.
